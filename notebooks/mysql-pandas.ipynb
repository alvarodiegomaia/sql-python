# Using MySQL in python

```python
import numpy as np
import pandas as pd

import mysql.connector
```

## Raw Connection to DataBase

```python
connection = mysql.connector.connect(
    user='root',
    password='college',
    host='localhost',
    database='employees',
    ssl_disabled=True
)

cursor = connection.cursor()
```

```python
query = '''
    SELECT *
    FROM employees e
    JOIN salaries s 
    ON e.emp_no  = s.emp_no 
    WHERE e.hire_date > "1999-01-01"
'''
```

```python
cursor.execute(query)
```

```python
results = []
for i, data in enumerate(cursor):
    results.append(data)

cursor.close()
connection.close()
```
